```python
#!/usr/bin/env python

'''
Multithreaded video processing sample.
Usage:
   video_threaded.py {<video device number>|<video file name>}
   Shows how python threading capabilities can be used
   to organize parallel captured frame processing pipeline
   for smoother playback.
Keyboard shortcuts:
   ESC - exit
   space - switch between multi and single threaded processing
'''


# Python 2/3 compatibility
from __future__ import print_function

import numpy as np
import cv2 as cv

from multiprocessing.pool import ThreadPool
from collections import deque

from common import clock, draw_str, StatValue
import video


rad = 30 #this is how big the circles are that represent the object. You will need to left click on the objects when the video starts.

obj1posSet = set()
obj2posSet = set()

obj1interacting_frames = []
obj2interacting_frames = []

head_pos = []
midhead_pos = []
fieldCorners = set()

obj1interaction_frames = 0
obj2interaction_frames = 0
obj1indicator = int(20),int(20) #where the interaction indicator is
obj2indicator = int(70),int(20)

class DummyTask:
    def __init__(self, data):
        self.data = data
    def ready(self):
        return True
    def get(self):
        return self.data

def main():
    import sys

    try:
        fn = filename
    except:
        fn = 0
    cap = video.create_capture(fn)

    
    def objectBorder(event, x, y, flags, param):
       global obj1posSet, obj2posSet

       if event == cv.EVENT_LBUTTONDOWN:
            obj1posSet.add((x, y))
       if event == cv.EVENT_RBUTTONDOWN:
            obj2posSet.add((x, y))
       if event ==cv.EVENT_MBUTTONDOWN:
            fieldCorners.add((x, y))
        
    def distance(point1, point2):
        x_dist = abs(point1[0] - point2[0])
        y_dist = abs(point1[1] - point2[1])
        return math.hypot(x_dist,y_dist)
        
    def process_frame(frame, t0, frame_num):
        # some intensive computation...
#         frame = cv.medianBlur(frame, 19)
#         frame = cv.medianBlur(frame, 19)
        global head_dir
    
        xr = rightEar[frame_num][0]
        yr = rightEar[frame_num][1]
        xl = leftEar[frame_num][0]
        yl = leftEar[frame_num][1]
        xs = snout[frame_num][0]
        ys = snout[frame_num][1]
        
        midxmid = (xl+xr+xs)/3.0 #midpoint between snout and ears
        midymid = (yl+yr+ys)/3.0
        
        xmid = (xl+xr)/2.0 #midpoint between just the ears
        ymid = (yl+yr)/2.0
        midhead_mid = int(np.round(midxmid)),int(np.round(midymid)) #middle of the head as midpoint between snout and ears
        head_mid = int(np.round(xmid)),int(np.round(ymid)) #define middle of the head as midpoint between left and right corners of implant
        tailx = tail[frame_num][0]
        taily = tail[frame_num][1]
        
        gap = 10.0 #how far in front/behind of head mid the point representing head direction will be.
        slope = ((yr-yl)/(xr-xl)) #slope between left and right corners of implant
        negx = 0
        posx = 0
        negy = 0
        posy = 0
        head_dir = 0
        
        dy = math.sqrt(gap**2/(1+slope**2)) #use pythagorean theorem to find x and y displacement
        dx = -slope*dy
        posx = xmid + dx
        posy = ymid + dy
        negx = xmid - dx
        negy = ymid - dy
           
        poshead_dir = posx, posy #two solutions to square root
        neghead_dir = negx, negy
        
        
        if snoutprob[frame_num] > 0.95: #if the snout reading is accurate, use that as head dir
            head_dir = snout[frame_num]
        elif rightprob[frame_num] > 0.95 and leftprob[frame_num] > 0.95:
            if math.sqrt((posx-tailx)**2+(posy-taily)**2) > math.sqrt((negx-tailx)**2+(negy-taily)**2): #otherwise use midpoint perpendicular to slope btwn implant corners
                head_dir = poshead_dir
            else:
                head_dir = neghead_dir
        else:
            head_dir = 0,0
        
        head_pos.append(head_dir)
        midhead_pos.append(midheadmid)
        drawTailCircleHere = int(np.round(tail[frame_num][0])),int(np.round(tail[frame_num][1]))
        drawleftEarCircleHere = int(np.round(leftEar[frame_num][0])),int(np.round(leftEar[frame_num][1]))
        drawrightEarCircleHere = int(np.round(rightEar[frame_num][0])),int(np.round(rightEar[frame_num][1]))
        drawsnoutCircleHere = int(np.round(snout[frame_num][0])),int(np.round(snout[frame_num][1]))
        
        drawhead_directionCircleHere = int(np.round(head_dir[0])), int(np.round(head_dir[1]))
                                 
        #frame = cv.circle(frame, drawledCircleHere, 4, (255, 100, 0), 2)
        #frame = cv.circle(frame, drawTailCircleHere, 4, (0, 100, 200), 2)
        #frame = cv.circle(frame, drawBodyCircleHere, 4, (0, 100, 200), 2)
        #frame = cv.circle(frame, drawmidCircleHere, 4, (50, 50, 0), 2)
        #frame = cv.circle(frame, drawbackCircleHere, 4, (0, 50, 75), 2)
        #frame = cv.circle(frame, drawleftCircleHere, 4, (0, 255, 255), 2)
        #frame = cv.circle(frame, drawrightCircleHere, 4, (100, 100, 100), 2)
        #frame = cv.circle(frame, head_mid, 4, (200,200,100), 2)
        frame = cv.circle(frame, drawhead_directionCircleHere, 4, (0,255,0), 2)
        #frame = cv.circle(frame, drawsnoutCircleHere, 4, (0, 20, 255), 2)
        
        for point in obj1posSet:
            drawPointHere = int(np.round(point[0])),int(np.round(point[1])) #make the labeled border points into integers
            cv.circle(frame, drawPointHere, rad, (255,255,255), 1) #define a circle at each point for visualization purposes
            
            if distance(head_dir,point) < rad: #if the head direction point is close enough to the labeled point, it's interacting
                global obj1interaction_frames #need global because otherwise it won't save between frames
                obj1interaction_frames += 1
                cv.circle(frame, obj1indicator, 20, (0,255,0), -1) #make the indicator light green
                obj1interacting_frames.append(frame_num) #add the frame number to the list of interacting frames
                break #if interacting with one point, then don't need to check the rest
                
            else:
                cv.circle(frame, obj1indicator, 20, (0,0,255), -1) #if not interacting, make indicator light red
        #cv.imwrite("images/%04d.tiff"%frame_num,frame)
        for point in obj2posSet:
            drawPointHere = int(np.round(point[0])),int(np.round(point[1])) 
            cv.circle(frame, drawPointHere, rad, (255,255,255), 1) 
            
            if distance(head_dir,point) < rad:
                global obj2interaction_frames
                obj2interaction_frames += 1
                cv.circle(frame, obj2indicator, 20, (0,255,0), -1) #make the indicator light green
                obj2interacting_frames.append(frame_num)
                break #if interacting with one point, then don't need to check the rest
                
            else:
                cv.circle(frame, obj2indicator, 20, (0,0,255), -1) #if not interacting, make indicator light red
                
        for point in fieldCorners:
            drawPointHere = int(np.round(point[0])),int(np.round(point[1]))
            cv.circle(frame, drawPointHere, 3, (255,255,255), -1)
            
        #cv.imwrite("images/%04d.tiff"%frame_num,frame)
        return frame, t0
         
    threadn = cv.getNumberOfCPUs()
    pool = ThreadPool(processes = threadn)
    pending = deque()
    
    threaded_mode = True

    latency = StatValue()
    frame_interval = StatValue()
    last_frame_time = clock()
    framenum = 0
    
    while framenum < len(leftEar):
        while len(pending) > 0 and pending[0].ready():
            res, t0 = pending.popleft().get()
            latency.update(clock() - t0)
            draw_str(res, (20, 20), "threaded      :  " + str(threaded_mode))
            draw_str(res, (20, 40), "latency        :  %.1f ms" % (latency.value*1000))
            draw_str(res, (20, 60), "frame interval :  %.1f ms" % (frame_interval.value*1000))
            cv.namedWindow('test',cv.WINDOW_AUTOSIZE)
            #cv.moveWindow('test',0,0)
            cv.imshow('test',res)
            #cv.imshow('threaded video', res)
            
        if len(pending) < threadn:
            
            cv.setMouseCallback('test', objectBorder)
            _ret, frame = cap.read()
            framenum += 1
            t = clock()
            frame_interval.update(t - last_frame_time)
            last_frame_time = t
            
            if threaded_mode:
                task = pool.apply_async(process_frame, (frame.copy(), t,framenum))
            else:
                task = DummyTask(process_frame(frame, t,framenum))
            pending.append(task)
            
        ch = cv.waitKey(1)
        
        if ch == ord(' '):
            threaded_mode = not threaded_mode
        if ch == 27:
            obj1posSet.clear()
            obj2posSet.clear()
            break
            
       
    obj1PosArray = np.array(list(obj1posSet))
    obj2PosArray = np.array(list(obj2posSet))
    fieldCornersArray = np.array(list(fieldCorners))
    headPos = np.array(head_pos)
    midheadPos = np.array(midhead_pos)
    radArray = np.asarray(rad)
    obj1IntFrameArray = np.array(obj1interacting_frames)
    obj2IntFrameArray = np.array(obj2interacting_frames)
    obj1NumofIntFrames = np.array(obj1interaction_frames)
    obj2NumofIntFrames = np.array(obj2interaction_frames)
    totalFrames = np.asarray(len(leftEar))
    print('Done')
    print("Seconds spent interacting with object 1:", obj1interaction_frames/60)
    print("Seconds spent interacting with object 2:", obj2interaction_frames/60)
    print("interacting frames:", obj1interaction_frames, obj2interaction_frames)
    print(fieldCorners)
    np.savez(npsavezpath, obj1NumofIntFrames = obj1NumofIntFrames, obj2NumofIntFrames = obj2NumofIntFrames, headPos = headPos, midheadPos = midheadPos, obj1PosArray = obj1PosArray,obj2PosArray = obj2PosArray, fieldCornersArray = fieldCornersArray, radArray = radArray, obj1IntFrameArray = obj1IntFrameArray, obj2IntFrameArray = obj2IntFrameArray, totalFrames = totalFrames)
    
if __name__ == '__main__':
    time = clock()
    print(__doc__)
    filename = f"/Volumes/JimDrive/{rat}/{data}.mp4"
    #headcoords_on = np.load("demo_headcoords_on.npz")['headcoords_on']
    main()
    cv.destroyAllWindows()
    print("This took %6.2f seconds"%(clock() - time))
    
    
    
```
