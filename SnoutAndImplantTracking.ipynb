```python
#!/usr/bin/env python

'''
Multithreaded video processing sample.
Usage:
   video_threaded.py {<video device number>|<video file name>}
   Shows how python threading capabilities can be used
   to organize parallel captured frame processing pipeline
   for smoother playback.
Keyboard shortcuts:
   ESC - exit
   space - switch between multi and single threaded processing
'''


# Python 2/3 compatibility
from __future__ import print_function

import numpy as np
import cv2 as cv

from multiprocessing.pool import ThreadPool
from collections import deque

from common import clock, draw_str, StatValue
import video

NUMFRAMES = 1525
filename = "/Users/jinjiang-macair/Desktop/DLC/SnoutAndImplantTracking-Jim-2020-05-20/videos/test copy 3.mp4"
posSet = {}
interaction_frames = 0


class DummyTask:
    def __init__(self, data):
        self.data = data
    def ready(self):
        return True
    def get(self):
        return self.data

def main():
    import sys

    try:
        fn = filename
    except:
        fn = 0
    cap = video.create_capture(fn)

    
    def objectBorder(event, x, y, flags, param):
       global posSet
       if event == cv.EVENT_LBUTTONDOWN:
            posSet.add((x, y))
           
    def distance(point1, point2):
        x_dist = abs(point1[0] - point2[0])
        y_dist = abs(point1[1] - point2[1])
        return math.hypot(x_dist,y_dist)
        
    def process_frame(frame, t0, frame_num):
        # some intensive computation...
#         frame = cv.medianBlur(frame, 19)
#         frame = cv.medianBlur(frame, 19)
        global head_dir
        xr = right[frame_num][0]
        yr = right[frame_num][1]
        xl = left[frame_num][0]
        yl = left[frame_num][1]
        xmid = (xl+xr)/2.0
        ymid = (yl+yr)/2.0
        head_mid = int(np.round(xmid)),int(np.round(ymid))
        
        tailx = tail[frame_num][0]
        taily = tail[frame_num][1]
        
        gap = 10.0
        slope = ((yr-yl)/(xr-xl))
        negx = 0
        posx = 0
        negy = 0
        posy = 0
        head_dir = 0
        
        
        dy = math.sqrt(gap**2/(1+slope**2))
        dx = -slope*dy
        posx = xmid + dx
        posy = ymid + dy
        negx = xmid - dx
        negy = ymid - dy
           
        poshead_dir = posx, posy
        neghead_dir = negx, negy
        
        if snoutprob[frame_num] > 0.95:
            head_dir = snout[frame_num]
        elif math.sqrt((posx-tailx)**2+(posy-taily)**2) > math.sqrt((negx-tailx)**2+(negy-taily)**2):
            head_dir = poshead_dir
        else:
            head_dir = neghead_dir
            
        drawledCircleHere = int(np.round(led[frame_num][0])),int(np.round(led[frame_num][1]))
        drawTailCircleHere = int(np.round(tail[frame_num][0])),int(np.round(tail[frame_num][1]))
        drawleftCircleHere = int(np.round(left[frame_num][0])),int(np.round(left[frame_num][1]))
        drawrightCircleHere = int(np.round(right[frame_num][0])),int(np.round(right[frame_num][1]))
        drawsnoutCircleHere = int(np.round(snout[frame_num][0])),int(np.round(snout[frame_num][1]))
        #drawmidCircleHere = int(np.round(mid[frame_num][0])),int(np.round(mid[frame_num][1]))
        drawBodyCircleHere = int(np.round(body[frame_num][0])),int(np.round(body[frame_num][1]))
        #drawbackCircleHere = int(np.round(back[frame_num][0])),int(np.round(back[frame_num][1]))
        drawhead_directionCircleHere = int(np.round(head_dir[0])), int(np.round(head_dir[1]))
                                 
        #frame = cv.circle(frame, drawledCircleHere, 4, (255, 100, 0), 2)
        #frame = cv.circle(frame, drawTailCircleHere, 4, (0, 100, 200), 2)
        #frame = cv.circle(frame, drawBodyCircleHere, 4, (0, 100, 200), 2)
        #frame = cv.circle(frame, drawmidCircleHere, 4, (50, 50, 0), 2)
        #frame = cv.circle(frame, drawbackCircleHere, 4, (0, 50, 75), 2)
        #frame = cv.circle(frame, drawleftCircleHere, 4, (0, 255, 255), 2)
        #frame = cv.circle(frame, drawrightCircleHere, 4, (100, 100, 100), 2)
        #frame = cv.circle(frame, head_mid, 4, (200,200,100), 2)
        frame = cv.circle(frame, drawhead_directionCircleHere, 4, (0,255,0), 2)
        #frame = cv.circle(frame, drawsnoutCircleHere, 4, (0, 20, 255), 2)
        
        #cv.imwrite("images/%04d.tiff"%frame_num,frame)
        return frame, t0
         
    threadn = cv.getNumberOfCPUs()
    pool = ThreadPool(processes = threadn)
    pending = deque()
    
    threaded_mode = True

    latency = StatValue()
    frame_interval = StatValue()
    last_frame_time = clock()
    framenum = 0
    rad = 10
    while framenum < len(left):
        while len(pending) > 0 and pending[0].ready():
            res, t0 = pending.popleft().get()
            latency.update(clock() - t0)
            draw_str(res, (20, 20), "threaded      :  " + str(threaded_mode))
            draw_str(res, (20, 40), "latency        :  %.1f ms" % (latency.value*1000))
            draw_str(res, (20, 60), "frame interval :  %.1f ms" % (frame_interval.value*1000))
            cv.imshow('threaded video', res)
            
        if len(pending) < threadn:
            _ret, frame = cap.read()
            framenum += 1
            t = clock()
            frame_interval.update(t - last_frame_time)
            last_frame_time = t
            cv.setMouseCallback('threaded video', objectBorder)
            posNp = np.array(posList)
            
            if threaded_mode:
                task = pool.apply_async(process_frame, (frame.copy(), t,framenum))
            else:
                task = DummyTask(process_frame(frame, t,framenum))
                
            for point in posList:
                drawPointHere = int(np.round(point[0])),int(np.round(point[1]))
                frame = cv.circle(frame, drawPointHere, 4, (255,255,255), 2)
                if distance(head_dir,point) < rad:
                    global interaction_frames
                    interaction_frames += 1
                    
            pending.append(task)
        ch = cv.waitKey(1)
        if ch == ord(' '):
            threaded_mode = not threaded_mode
        if ch == 27:
            posSet.clear()
            break
        print(interaction_frames)
        print(posSet)
    print('Done')


if __name__ == '__main__':
    time = clock()
    print(__doc__)
    filename = "/Users/jinjiang-macair/Desktop/DLC/SnoutAndImplantTracking-Jim-2020-05-20/videos/test copy 3.mp4"
    #headcoords_on = np.load("demo_headcoords_on.npz")['headcoords_on']
    main()
    cv.destroyAllWindows()
    print("This took %6.2f seconds"%(clock() - time))
```
